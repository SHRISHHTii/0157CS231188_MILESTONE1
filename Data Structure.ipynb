Dictionary

```python
d = {0: 10, 1: 20}

d[2] = 30

print(d)
```

```python
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

result = {}

result.update(dic1)
result.update(dic2)
result.update(dic3)

print(result)
```

```python
d = {1: "Apple", 2: "Banana", 3: "Mango"}

key = int(input("Enter key: "))

if key in d:
    print("Key exists")
else:
    print("Key does not exist")
```

```python
d = {
    "Name": "John",
    "Age": 21,
    "City": "Delhi"
}

print("Keys:")
for key in d.keys():
    print(key)

print("\nValues:")
for value in d.values():
    print(value)

print("\nKeys and Values:")
for key, value in d.items():
    print(key, ":", value)
```

```python
d = {}

for i in range(1, 16):
    d[i] = i * i

print(d)
```

```python
d = {
    "a": 100,
    "b": 200,
    "c": 300
}

total = sum(d.values())

print("Sum =", total)
```

```python
set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

result = set1.intersection(set2)

print("Intersection:", result)
```

```python
set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

result = set1.union(set2)

print("Union:", result)
```

String

```python
s = input("Enter a string: ")

upper = 0
lower = 0

for ch in s:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase letters:", upper)
print("Lowercase letters:", lower)
```

```python
s = input("Enter a string: ")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
s = input("Enter a string: ")

result = s[:2] * len(s)

print(result)
```

```python
s = input("Enter a string: ")

if s.startswith('x'):
    s = s[1:]

if s.endswith('x'):
    s = s[:-1]

print(s)
```

```python
s = input("Enter a string: ")
n = int(input("Enter n: "))

result = s[-n:] * n

print(result)
```
