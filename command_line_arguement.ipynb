```python
like = input("Enter liked numbers: ").split("-")
dislike = input("Enter disliked numbers: ").split("-")
numbers = input("Enter numbers: ").split("-")

happiness = 0

for num in numbers:
    if num in like:
        happiness += 1
    elif num in dislike:
        happiness -= 1

print(happiness)
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

nums = list(map(int, input("Enter 10 numbers: ").split()))

total = 0

for num in nums:
    if is_prime(num):
        total += num

print("Sum of prime numbers =", total)
```
